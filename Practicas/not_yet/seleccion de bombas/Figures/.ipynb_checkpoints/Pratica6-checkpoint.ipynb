```python
import numpy as np
import scipy.optimize
```

```python
def cole_turb(e_d):
    return (-2*np.log10(e_d/3.7))**-2

def coleb(Re,e_d):
    eqf = lambda f : f**-0.5+2*np.log10(e_d/3.7+2.51/Re/f**0.5)
    ffact = scipy.optimize.newton(eqf,cole_turb(e_d))
    return ffact
```

```python
Ds = 102.3 *10**-3 # m
Ats = Ds**2*np.pi/4
Dd = 102.3 *10**-3 # m
Atd = Dd**2*np.pi/4

Ls = 2 # m
Ld = 400 # m

e = 4.6*10**-5 #m
e_dd = e/Dd # rugosidad relativa linea de descarga
e_ds = e/Ds # rugosidad relativa linea de succion

g=9.81 # m/s2
mu = 10**-3 # Pa.s
rho = 10**3 # kg/m3
Q = 800*10**-3/60 # m3/s
za = 0
zb = 80 # m

def ha_func(q):
    
    vd = q/Atd
    vs = q/Ats
    
    # perdidas de carga mayores
    Res = rho*vs*Ds/mu
    fs  = coleb(Res,e_ds)
    hls = fs*Ls/Ds*vs**2/2/g 
    
    Red = rho*vd*Dd/mu
    fd  = coleb(Red,e_dd)
    hld = fd*Ld/Dd*vd**2/2/g 
    
    #total
    hl = hld+hls
    
    # perdidas de carga menores
    #succion
    fts = cole_turb(e_ds)
    k1  = 75*fts 
    k2  = 30*fts
    
    hlms = vs**2/2/g*(k1+k2)

    #descarga
    ftd = cole_turb(e_dd)
    k3  = 30*ftd
    k4  = 100*ftd
    k5  = 8*ftd
    k6  = 30*ftd
    k7  = 1

    hlmd = vd**2/2/g*(k3+k4+k5+k6+k7)
    
    #total
    hlm  = hlms+hlmd
    
    ha   = zb-za+hl+hlm
    print(hlm,hl)
    return ha

ha_req = ha_func(Q)
print('ha requerida =',ha_req)
```

```python
Ds = 62.7 *10**-3 # m
Ats = Ds**2*np.pi/4
Dd = 102.3 *10**-3 # m
Atd = Dd**2*np.pi/4
Dd1 = 52.5*10**-3 # m
Atd1 = Dd1**2*np.pi/4

Ls = 2 # m
Ld = 400 # m

e = 4.6*10**-5 #m
e_dd = e/Dd # rugosidad relativa linea de descarga
e_ds = e/Ds # rugosidad relativa linea de succion

g=9.81 # m/s2
mu = 10**-3 # Pa.s
rho = 10**3 # kg/m3
Q = 800*10**-3/60 # m3/s
za = 0
zb = 80 # m

def ha_func(q):
    
    vd = q/Atd
    vs = q/Ats
    vd1 = q/Atd1
    
    # perdidas de carga mayores
    Res = rho*vs*Ds/mu
    fs  = coleb(Res,e_ds)
    hls = fs*Ls/Ds*vs**2/2/g 
    
    Red = rho*vd*Dd/mu
    fd  = coleb(Red,e_dd)
    hld = fd*Ld/Dd*vd**2/2/g 
    
    #total
    hl = hld+hls
    
    # perdidas de carga menores
    #succion
    fts = cole_turb(e_ds)
    k1  = 75*fts 
    k2  = 30*fts
    
    hlms = vs**2/2/g*(k1+k2)

    #descarga
    ftd = cole_turb(e_dd)
    k3  = 30*ftd
    k4  = 100*ftd
    k5  = 8*ftd
    k6  = 30*ftd
    k7  = 1
    k8  = 0.46

    hlmd = vd**2/2/g*(k3+k4+k5+k6+k7)+vd1**2/2/g*k8
    
    #total
    hlm  = hlms+hlmd
    
    ha   = zb-za+hl+hlm

    return ha

ha_req = ha_func(Q)
print('ha requerida =',ha_req)
```

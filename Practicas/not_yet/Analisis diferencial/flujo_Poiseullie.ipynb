```python
import numpy as np
```

```python
r_R=np.linspace(0,1,1000)
vz_vmax=(1-np.square(r_R))
tau_tauR=(r_R)
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(6,6),dpi=300)
ax = fig.add_subplot(111)

ax.plot(r_R,vz_vmax)
ax.set_xlabel(r'$r/R$',fontsize=18)
ax.set_ylabel(r'$v_z/v_{z,max}$',fontsize=18)
ax.set_xlim(0,1)
ax.set_ylim(0,1)
```

```python
fig = plt.figure(figsize=(6,6),dpi=300)
ax = fig.add_subplot(111)

ax.plot(r_R,tau_tauR)
ax.set_xlabel(r'$r/R$',fontsize=18)
ax.set_ylabel(r'$\tau_{rz}/\tau_{rz,pared}$',fontsize=18)
ax.set_xlim(0,1)
ax.set_ylim(0,1)
```

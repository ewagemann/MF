```python
import sympy as sy
```

```python
r,R1,R2,C1,C2,mu,omega=sy.symbols('r,R1,R2,C1,C2,mu,omega')
```

```python
sy.integrate(1/r**(3))
```

```python
exp1 = C1/2/mu/R1+C2*R1-omega*R1
exp2 = C1/2/mu/R2+C2*R2

sol=sy.solve([exp1,exp2],[C1,C2])
sol[C1]
```

```python
sol[C2]
```

```python
vtheta=sy.simplify(sol[C1]/2/mu/r+sol[C2]*r)
```

```python
vtheta
```

```python
v_av=sy.simplify(sy.integrate(vtheta,[r,R1,R2])/(R2-R1))
```

```python
v_av
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
R1=0.25
R2=1
r=np.linspace(R1,R2,1000)

vtheta=(R2/r-r/R2)/(R2/R1-R1/R2)
fig=plt.figure(figsize=(6,6),dpi=300)
ax=fig.add_subplot(111)
ax.plot(r,vtheta)
rect = patches.Rectangle((0, 0), R1, 1, linewidth=1, edgecolor='lime', facecolor='none', hatch="/")
ax.add_patch(rect)

ax.plot([R1,R1],[0,1],'k--')

ax.set_xlabel(r'$r/R_2$',fontsize=18)
ax.set_ylabel(r'$v_\theta/(\omega R_1)$',fontsize=18
             )

ax.set_xlim(0,1)
ax.set_ylim(0,1)

```

```python
R1=0.25
R2=1
r=np.linspace(R1,R2,1000)

omega=(R2/r-r/R2)/(R2/R1-R1/R2)/r*R1
fig=plt.figure(figsize=(6,6),dpi=300)
ax=fig.add_subplot(111)
ax.plot(r,omega)
rect = patches.Rectangle((0, 0), R1, 1, linewidth=1, edgecolor='lime', facecolor='none', hatch="/")
ax.add_patch(rect)

ax.plot([R1,R1],[0,1],'k--')

ax.set_xlabel(r'$r/R_2$',fontsize=18)
ax.set_ylabel(r'$\omega$',fontsize=18
             )

ax.set_xlim(0,1)
ax.set_ylim(0,1)
```

```python
R1=0.25
R2=1
r=np.linspace(R1,R2,1000)

tau = R1**2/r**2

fig=plt.figure(figsize=(6,6),dpi=300)
ax=fig.add_subplot(111)
ax.plot(r,tau)
rect = patches.Rectangle((0, 0), R1, 1, linewidth=1, edgecolor='lime', facecolor='none', hatch="/")
ax.add_patch(rect)

ax.plot([R1,R1],[0,1],'k--')

ax.set_xlabel(r'$r/R_2$',fontsize=18)
ax.set_ylabel(r'$\tau_{r\theta}/\tau_{r\theta,\,pared\, 1}$',fontsize=18
             )

ax.set_xlim(0,1)
ax.set_ylim(0,1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sigmaco = 3.38
epco = 4.664
sigmach = 2.7
epch = 2.487
```

```python
r = np.linspace(2,10,1000)

def LJ(ep,sigma):
    return np.multiply(4*ep,np.subtract(np.power(np.divide(sigma,r),12),np.power(np.divide(sigma,r),6)))
Uo = LJ(epco,sigmaco)
Uh = LJ(epch,sigmach)

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111)

ax.plot(r,Uo)
ax.plot(r,Uh)
ax.set_ylim(min(Uo),0)
```
